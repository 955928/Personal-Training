```python
import random

def game_on(number, random_num):
    #définit une valeur à mon variable
    number_of_tries = 0
    
    #création d'une boucle for qui ne cessera qu'à certaine condition réalisé par l'utilisateur
    while True:
        #defintion d'une valeur à ma variable 
        user_input = input("Enter a number to guess the right one: ")
        #création d'une condition qui une fois rempli cesse le fonctionnement du programme;si utilisateur entre rien 
        if not user_input:
            break
        #conversion de ma variable de str() en int()
        user_input = int(user_input)
        #condition si input est entre 1 et 100
        if 1 <= user_input <= 100:
            #compteur pour le nombre d'essais de l'utilisateur
            number_of_tries += 1
            #condition si input est le nombre aléatoire génnéré par le système qui résulte dans sa réussite
            if user_input == random_num:
                #notification de la réusssite de l'utilisateur
                return f"Bravo! You've won the game. It took you {number_of_tries} tries to win."
            #si input est moins que nombre aléatoire
            elif user_input < random_num:
                #notification au utilisateur que l'input est trop petit
                print(f"Sorry, the number's too small! Try again! This is your {number_of_tries} attempt.")
            else:
                #notification que l'input est trop grand
                print(f"Sorry, the number's too large! Try again! This is your {number_of_tries} attempt.")
        else:
            #si l'input n'est pas entre les limites de la condition if
            return "Error: Number submitted must be between 1 and 100"


    
   
#Création d'une fonction pour l'option facile(l'explication serait la même pour les deux autres difficultés)
def Easy():
    #je crée un print qui acceuil l'utilisateur et l'explique les instruction pour le jeu   
    print("You have chosen Easy.You are only allowed to choose any number between 1 - 30")
    #je fixe la limite dans laquelle la fonction random va modifier
    number_range = (1, 30)
    #fonction quui permet de produire le nombre aléatoire choisi par le systeme dans la limite de la num_range 
    random_num = random.randint(*number_range)
    #je retourne la fonction principale dans laquelle va se trouver cela(def Easy())
    return game_on(number_range, random_num)



#création d'une fonction pour  l'option intermédiaire
def Intermediate():
    print("You have chosen Intermediate.You are only allowed to choose any number between 1-60")
    number_range = (1, 60)
    random_num = random.randint(*number_range)
    return game_on(number_range, random_num)


#création d'une fonction pour l'option difficile     
def Difficult(): 
    print("You have chosen Difficult.You are only allowed to choose any number between 1 - 100")
    number_range = (1, 100)
    random_num = random.randint(*number_range)
    return game_on(number_range, random_num)  




choose_difficulty = input("Choose the difficulty you wish: (Difficult, Intermediate, Easy)")

if choose_difficulty in ["Easy", "Intermediate", "Difficult"]:
    result = globals()[choose_difficulty]()
    print(result)
else:
    print("Invalid entry. Please choose among the provided options.")
```

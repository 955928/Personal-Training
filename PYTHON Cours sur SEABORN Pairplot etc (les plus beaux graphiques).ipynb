```python
#"load_datatset" = fonction qui  permet de charger des dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.pairplot(titanic)
#résultat dû au fait dans ce dataset on a bcp de variables qui forment des catégories
```

```python
#"catplot"

sns.catplot(x='pclass', y='age', data=titanic)
```

```python


sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
#les "boxplot" = des graphiques qui nous permettent de voir la répartition des données
#avec les principales statistiques,les quartis, les médianes etc sur différents catégories de variables
#on retrouve toujours la médiane d'une série de donnée au milieu d'une boxplot
#aux extrémités le premier quartile et le 3eme quartile
#les traits qui allongent les boites(les whiskers) nous montrent toute l'étendue des valeurs restantes est égale à 0,5 fois la différence entre le quartile 1 et 3
#on retrouves les anomalies qu'on pourra trouver dans nos données

sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
# "distplot"= visulaiser la dirtibution de l'élément visé ici "fare"

sns.distplot(titanic['fare'])
```

```python
#"jointplot" = visualiser la variable de deux distribution à la fois

sns.jointplot(x='age', y='fare', data=titanic)
```

```python
#"kind" = créer un style dans le graphique
sns.jointplot(x='age', y='fare', data=titanic, kind='kde')#karnal density
```

```python
#"kind" = créer un style dans le graphique
sns.jointplot(x='age', y='fare', data=titanic, kind='hex')#hexagonal
```

```python
#"heatmap" = visualiser des matrices, surotut des matrices de coorélation

sns.heatmap(titanic.corr())
```

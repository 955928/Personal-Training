```python
import numpy as np

x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
#tracer un nuage de points
plt.scatter(x,y)
```

```python
#Customiser une courbe
plt.plot(x, y, c='red')
```

```python
#linewidth
plt.plot(x, y, c='red',lw=5)
```

```python
#linestyle
plt.plot(x, y, c='red',lw=5, ls='--')
```

```python
#Le cylce d'une vie d'une figure
#normalement on commence avec plt.figure(juste une feuille de travail sur laquelle on va mettre nos donées)
plt.figure()
```

```python
#on va pouvoir chosiir la taille de notre graphique; en inch et pas en cm
plt.figure()
plt.plot(x,y, label='quadratique' )
#on paut rajouter d'autre fiugure sur notre graphique
plt.plot(x, x**3, label='cubique')
plt.title('figure1')
plt.xlabel('axe x')
plt.ylabel('axe y')
#definir un légénde, donc on l'écrivant dans les deux codes pour leur courbes respectives
plt.legend()
#enregistrer un image de notre graphique dans notre répertoire de travail
plt.savefig('figure.png')

#ailleurs que dans jupyter  pour créeer une autre figure on repète le même syntaxe plus loins dans le code initial alors qu'avec jupyter on a que à aller à une nouvelle ligne de code
#plt.figure()
#plt.plot(x,y)
#etc
```

```python
#subplot permet de générer une grille de different graphique sur notre figure
#créer 2 graphqiue sur une seule colonne
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
#pour obtenir le graphique qui se situe en dessous
```

```python

plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
#pour obtenir le graphique qui se situe en dessous
plt.subplot(2, 1, 2)
plt.plot(x, y, c='blue')
```

```python
#Une véritable cycle de vie d'une figure en MATPLOTLIB ressemble à ceci
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Graphique 1')
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.title("Graphique 2")#I added this just top try out something
plt.legend()
```

```python
#Autre méthode: OOP (Orienté Objet) plus poussé mais moins recommandé parceque avec la methode plt.figure  on peut tracer 99% des graphiques désirés
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
#on a plus de possibilités en OOP , on peut créer plusieursgraphique qui ont la même ordonée
#faut écrire dans le plt.subplots()
fig, ax = plt.subplots(2, 1, sharex=True)
ax.plot(x, y)
```

```python
#lorsqu'on a un subplots() qu'on enregistre dans la vairable AX, AX n'est pas une OBJET  mais un tableau numpy qui va contenir nos différents objets
fig, ax = plt.subplots(2, 1, sharex=True)
type(ax)
ax.shape
#tableau de 2 dimension
```

```python
#Donc la résolution au soucis
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
```
